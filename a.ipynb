```python
import cv2

cap = cv2.VideoCapture(0)  # ลองเปลี่ยนเป็น 1 หรือ 2 หาก 0 ไม่ได้ผล

if not cap.isOpened():
    print("Error: ไม่สามารถเปิดกล้องได้")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: ไม่สามารถรับเฟรมได้")
        break

    cv2.imshow('Camera Test', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
